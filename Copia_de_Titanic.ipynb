```python
import pandas as pd
from tabulate import tabulate
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('titanic-train.csv')
df.info()
```

```python
df.head(10)
```

```python
features = {
    "Feature": ["PassengerId", "Survived", "Pclass", "Name", "Sex", "Age", "Fare", "Embarked"],
    "Tipo de Dato": ["Numérico Discreto", "Categoríco Binario", "Numérico Ordinal",
                     "Texto", "Categoríco", "Numérico Continuo", "Numérico Continuo", "Categoríco"],
    "Descripción": ["Identificador único del pasajero",
                    "1 si sobrevivió, 0 si no",
                    "Clase del ticket (1, 2, 3)",
                    "Nombre del pasajero",
                    "Género del pasajero",
                    "Edad del pasajero",
                    "Tarifa pagada por el pasajero",
                    "Puerto de embarque (C, Q, S)"]
}
df_features = pd.DataFrame(features)
print(tabulate(df_features, headers="keys", tablefmt="fancy_grid"))
```

```python
colors = ["red", "blue"]
sns.countplot(data=df, x="Survived",hue="Survived", palette=colors, legend= False)
plt.xlabel("Supervivencia")
plt.ylabel("Cantidad de pasajeros")
plt.title("Supervivientes vs No sobrevivientes")
plt.show()
```

```python
colors2=["red","green","blue"]
sns.countplot(data=df, x="Pclass", hue="Pclass", palette=colors2, legend=False)
plt.xlabel("Clase del pasajero")
plt.ylabel("Cantidad de pasajeros")
plt.title("Distribución de pasajeros por clase")
plt.show()
```

```python
sns.countplot(data=df, x="Sex", hue="Sex", palette=colors, legend=False)
plt.xlabel("Género del pasajero")
plt.ylabel("Cantidad de pasajeros")
plt.title("Distribución de pasajeros por género")
plt.show()
```

```python
df.isnull().sum()
```

```python
sns.histplot(data=df, x="Age", kde=True)
plt.xlabel("Edad del pasajero")
plt.ylabel("Frecuencia")
plt.title("Distribución de edades de los pasajeros")
plt.show()
```
