2. Convertir en dataframe el siguiente conjunto de medidas
```
medida_2020={"Ag":2,"Au":5,"Cu":3,"Pt:2}
medida_2021={"Ag":4,"Au":3,"Cu":6,"Pt:1}
medida_2022={"Ag":8,"Au":2,"Cu":15,"Pt:0}
medida_2023={"Ag":5,"Au":1,"Cu":0,"Pt:5}
medida_2024={"Ag":7,"Au":0,"Cu":3,"Pt:2}
medida_20205={"Ag":9,"Au":1,"Cu":3,"Pt:5}
```
Tenga en cuenta que los registros son los meses de medida y las features deben ser los elementos





````python
import pandas as pd

# Diccionario de elementos
elementos = {
    "número atómico": [1, 6, 47, 88],
    "Masa atómica": [1, 12, 107, 226],
    "Familia": ["No metal", "No metal", "Metal", "Metal"]
}

# Convertir el diccionario en un DataFrame
df_elementos = pd.DataFrame(elementos)

# Agregar nombres de los elementos como registros (nombres)
df_elementos["Elemento"] = ["Hidrógeno", "Carbono", "Plata", "Radio"]

# Cambiar el orden de las columnas para que "Elemento" sea la primera columna
df_elementos = df_elementos[["Elemento", "número atómico", "Masa atómica", "Familia"]]

# Mostrar el DataFrame resultante
print(df_elementos)
````

````python
df_elementos[["Elemento", "número atómico"]]
````

````python
df_elementos.dtypes
````

````python
# Diccionarios con medidas para diferentes años
medida_2020 = {"Ag": 2, "Au": 5, "Cu": 3, "Pt": 2}
medida_2021 = {"Ag": 4, "Au": 3, "Cu": 6, "Pt": 1}
medida_2022 = {"Ag": 8, "Au": 2, "Cu": 15, "Pt": 0}
medida_2023 = {"Ag": 5, "Au": 1, "Cu": 0, "Pt": 5}
medida_2024 = {"Ag": 7, "Au": 0, "Cu": 3, "Pt": 2}
medida_20205 = {"Ag": 9, "Au": 1, "Cu": 3, "Pt": 5}

# Convertir en DataFrame
df_medidas = pd.DataFrame({
    "2020": medida_2020,
    "2021": medida_2021,
    "2022": medida_2022,
    "2023": medida_2023,
    "2024": medida_2024,
    "20205": medida_20205
})

# Transponer el DataFrame para que los años sean los registros y los elementos sean las columnas
df_medidas = df_medidas.T

# Mostrar el DataFrame resultante
print(df_medidas)
````

````python
df_medidas[["Ag","Au"]]
````
